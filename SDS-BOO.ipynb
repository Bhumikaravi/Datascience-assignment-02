```python
# Univariate Analysis: Numerical
sns.histplot(df['Product Availability'], kde=True).set_title('Product Availability Distribution')
plt.show()
```

```python
# Univariate Analysis: Categorical
sns.countplot(x='Price Range', data=df).set_title('Price Range Count')
plt.show()
```

```python
# Univariate Pie Chart: Price Range distribution
gender_counts = df['Price Range'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Price Range Distribution')
plt.show()
```

```python
# Univariate Analysis: Categorical
sns.countplot(x='Order Type', data=df).set_title('Order Type Count')
plt.show()
```

```python
# Univariate Pie Chart: Product Availability distribution
depression_counts = df['Product Availability'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(depression_counts, labels=['Not in stock', 'Stock'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon'])
plt.title('Product Availability Distribution')
plt.show()
```

```python
# Univariate Analysis: Categorical
sns.countplot(x='Customer Feedback Type', data=df).set_title('Customer Feedback Type distribution')
plt.show()
```

```python
# Univariate Pie Chart:Customer Feedback Type
diet_counts = df['Customer Feedback Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Customer Feedback Type Distribution')
plt.show()
```

```python
#Agent Name vs Rating by scaatter plot
sns.scatterplot(x='Agent Name', y='Rating',hue='Order Type',  data=df, alpha=0.8)
plt.title('Agent Name vs Rating')
plt.xlabel('Agent Name')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
# Scatter plot colored by agent name vs Price Range
sns.scatterplot(x='Agent Name', y='Price Range', hue='Rating', data=df, alpha=0.7)
plt.title('agent name vs Price Range')
plt.xlabel('Agent Name')
plt.ylabel('Price Range')
plt.grid(True)
plt.show()
```

```python
#Agent name vs Rating scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Agent Name'], df['Rating'], color='teal', alpha=0.5)
plt.title('Agent name vs Rating')
plt.xlabel('Agent Name')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Discount Applied', hue='Price Range', data=df)
plt.title('Discount Applied by Price Range ')
plt.show()
```

```python
#Bivariate rating vs location
sns.boxplot(x='Rating', y='Location',hue='Price Range', data=df)
plt.title('Rating by location')
plt.show()
```

```python
# Scatter Plot (Bivariate):  colored by Product Availability
sns.scatterplot(x='Location', y='Rating', hue='Product Availability', data=df, alpha=0.9)
plt.title('Location vs rating')
plt.xlabel('Location')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Agent Name'], df['Location'], alpha=0.6, c='purple')
plt.title('agent name vs location')
plt.xlabel('Agent Name')
plt.ylabel('Location')
plt.grid(True)
plt.show()
```

```python
# Pairplot with Price Range hue  Multivariate Analysis
selected_columns = ['Delivery Time (min)','Rating','Customer Service Rating','Price Range']
sns.pairplot(df[selected_columns], hue='Price Range')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Correlation Heatmap
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
